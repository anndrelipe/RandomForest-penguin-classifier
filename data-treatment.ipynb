```python
# Dessas 3 últimas análises, infere-se que não existem nenhuma diferença substancialmente notável que difira os pinguins da
# espécie Adelie de diferentes ilhas. Dada a proximidade, já que o arquipélago de Palmer está só um pouco mais ao norte
# das ilhas Biscoe, é possível que haja especiação parapátrica:

'''
Especiação Alopátrica -->  Separadas por barreira geográfica, futuro isolamento reprodutivo e divergência genética;
' Simpátrica --> Não existem limitações geográficas mas também não ocorre cruzamento entre populações;
' Parapátricas --> Quando existe uma imposição geográfica mas não o suficiente para que não haja contado, possibilitando
                    reprodução entre populações.
'''

# Neste caso, a baixa diferenciação de nicho, visto que ocupam o norte da mesma península, reduz as divergências genéticas
# a curto prazo consequentemente a fenótica.

# Pensando assim, será elaborado dois modelos, um que diferencie apenas em relação as especíes, outro as espécies e ilha.
# Para ter uma primeira noção de qual será mais acertivo, farei a mesma avaliação entre diferentes espécies agora.
```
